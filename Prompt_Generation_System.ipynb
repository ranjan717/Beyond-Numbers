***Load the Transformer Model's Encoder***

```python
from transformers import AutoTokenizer, AutoModel
import torch
```

```python
tokenizer = AutoTokenizer.from_pretrained("distilbert-base-uncased")
model = AutoModel.from_pretrained("distilbert-base-uncased")
```

***Prepare a List of Sample User Queries***

```python
user_queries = [
    "What's the weather like today?",
    "Will it rain tomorrow?",
    "How hot will it be this weekend?",
    "Is there a chance of snow next week?",
    "What was the temperature yesterday?"
]
```

***Encode Each Query Using the Encoder Model***

```python
encoded_queries = []

for query in user_queries:
    inputs = tokenizer(query, return_tensors="pt")
    with torch.no_grad():
        outputs = model(**inputs)
    # We use the [CLS] token embedding as the sentence embedding
    encoded_query = outputs.last_hidden_state[:, 0, :].squeeze().numpy()
    encoded_queries.append(encoded_query)
```

***Generate Prompts for the AI to Respond To***

```python
prompts = [
    "The current weather is sunny with a high of 25°C.",
    "Yes, there's a 70% chance of rain tomorrow.",
    "It will be very hot this weekend, with temperatures reaching up to 35°C.",
    "Snow is unlikely next week, with temperatures expected to stay above freezing.",
    "Yesterday, the temperature was around 22°C."
]
```

***Display the Prompts and the Corresponding Encoded Queries for Verification***

```python
for query, encoded_query, prompt in zip(user_queries, encoded_queries, prompts):
    print(f"User Query: {query}")
    print(f"Generated Prompt: {prompt}")
    print("-" * 50)
```
