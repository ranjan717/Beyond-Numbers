```python
import torch
from transformers import MarianMTModel, MarianTokenizer
```

**English to Hindi Translation Model**


```python


# Load pre-trained MarianMT model and tokenizer for English to Hindi translation
model_name = 'Helsinki-NLP/opus-mt-en-hi'
tokenizer = MarianTokenizer.from_pretrained(model_name)
hindi_tran = MarianMTModel.from_pretrained(model_name)

# Input text
input_text = "Life is beautiful."
print("Input=",input_text)

# Tokenize input text
input_ids = tokenizer(input_text, return_tensors="pt")

# Perform translation
translated_output = hindi_tran.generate(**input_ids)

# Decode the translated output
translated_text = tokenizer.decode(translated_output[0], skip_special_tokens=True)

# Print the translated output
print("Translated Output (Hindi):")
print(translated_text)
```

**English to German Translation model**

**Create a model to choose the language option and then translate into that language.**

```python
# Load pre-trained MarianMT models and tokenizers for English to Hindi and English to German translation
models = {
    'Hindi': 'Helsinki-NLP/opus-mt-en-hi',
    'German': 'Helsinki-NLP/opus-mt-en-de'
}

tokenizers = {
    lang: MarianTokenizer.from_pretrained(model_name) for lang, model_name in models.items()
}

translation_models = {
    lang: MarianMTModel.from_pretrained(model_name) for lang, model_name in models.items()
}

def translate(text, target_language):
    if target_language not in models:
        raise ValueError(f"Translation to {target_language} is not supported.")

    tokenizer = tokenizers[target_language]
    model = translation_models[target_language]

    # Tokenize input text
    input_ids = tokenizer(text, return_tensors="pt")

    # Perform translation
    translated_output = model.generate(**input_ids)

    # Decode the translated output
    translated_text = tokenizer.decode(translated_output[0], skip_special_tokens=True)

    return translated_text

# User input
input_text = "Life is beautiful."
print("Input:", input_text)

# Choose target language
target_language = input("Enter target language (Hindi/German): ").strip()

# Translate and print the result
try:
    translated_text = translate(input_text, target_language)
    print(f"Translated Output ({target_language}):")
    print(translated_text)
except ValueError as e:
    print(e)
```
